***Customer Spending Analysis***

**Scenario:**

A retail store wants to analyze customer spending patterns.

**Tasks:**

Plot a Line Chart showing spending of a customer over 10 visits.

Use linewidth, color, and grid lines.

Use a Bar Chart to compare spending by different age groups (18–25, 26–35, 36–50, 50+).

Use horizontal bars (barh), colors, and labels.

Create a Pie Chart to visualize payment mode usage (Cash, Card, UPI).

Add autopct, shadow=True, and custom colors.

Use a Histogram to analyze distribution of average bill amounts.

Arrange line, bar, and pie charts in a subplot grid (2×2).

```python
import matplotlib.pyplot as plt
```

```python
visits=list(range(1,11))
spendings=[200, 250, 220, 300, 280, 350, 400, 370, 390, 420]
age_groups=["18-25","26-35","36-50","50+"]
# Assuming these values represent the total spending for each age group
age_group_spending = [5000, 7000, 8500, 6000]
payment_mode=["Cash", "Card", "UPI"]
payments = [2500, 5000, 3500]
bill_amounts=[200, 220, 250, 300, 280, 350, 400, 370, 390, 420,
                180, 210, 260, 300, 320, 330, 360, 410, 430, 450]
```

```python
plt.figure(figsize=(6,4))
plt.plot(visits, spendings, color="green", linewidth=2, marker="o", linestyle="--", label="Spending")
plt.xlabel("Visits")
plt.ylabel("Spending (in $)")
plt.title("Customer Spending Over 10 Visits")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.barh(age_groups, age_group_spending, color="gold")
plt.title("Customer Spending by Age Group")
plt.xlabel("Spending (in $)")
plt.ylabel("Age Group")
plt.show()
```

```python
plt.figure(figsize=(6,6))
colors = ["gold", "skyblue", "lightgreen"]
plt.pie(payments, labels=payment_mode, autopct="%1.1f%%", shadow=True, colors=colors, startangle=90)
plt.title("Payment Mode Usage")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(bill_amounts, bins=6, edgecolor="black", color="cyan")
plt.xlabel("Bill Amount ($)")
plt.ylabel("Frequency")
plt.title("Distribution of Average Bill Amounts")
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12,8))

# Line Chart
axs[0,0].plot(visits, spendings, color="blue", linewidth=2, marker="o", linestyle="--")
axs[0,0].set_title("Spending Over Visits")
axs[0,0].set_xlabel("Visits")
axs[0,0].set_ylabel("Spending ($)")
axs[0,0].grid(True)

# Barh Chart
axs[0,1].barh(age_groups, age_group_spending, color=["purple", "green", "orange", "red"])
axs[0,1].set_title("Age Group Spending")
axs[0,1].set_xlabel("Total Spending ($)")
axs[0,1].set_ylabel("Age Groups")

# Pie Chart
axs[1,0].pie(payments, labels=payment_mode, autopct="%1.1f%%", shadow=True, colors=["gold","skyblue","lightgreen"], startangle=90)
axs[1,0].set_title("Payment Mode Usage")

# Histogram
axs[1,1].hist(bill_amounts, bins=6, edgecolor="black", color="cyan")
axs[1,1].set_title("Bill Amount Distribution")
axs[1,1].set_xlabel("Bill Amount ($)")
axs[1,1].set_ylabel("Frequency")

plt.suptitle("Customer Spending Analysis", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
```
